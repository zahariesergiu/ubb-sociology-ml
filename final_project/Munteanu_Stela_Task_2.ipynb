```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
columns = [
    "age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
    "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
    "hours-per-week", "native-country", "income"
]

data = pd.read_csv(data_url, header=None, names=columns, na_values=" ?", skipinitialspace=True)
data.sample(10)
```

```python
data["workclass"].value_counts()
```

```python
data.replace("?", np.nan, inplace=True)
```

În acest dataset missing values sunt codate cu "?", care nu sunt recunoscute automat ca NaN

După replace se pot vedea explicit missing values in variabile

Acest pas este important pentru a asigura o analiză exploratorie accurate și o preprocesare corespunzătoare 

```python
missing_summary = pd.DataFrame({
    "Missing Count": data.isna().sum(),
    "Missing Percentage (%)": data.isna().mean() * 100
})

missing_summary
```

Variabilele cu lipsuri: workclass, occupation, native-country

Procent redus (<6%), dar relevant

```python
plt.style.use("default")

# Histogram + KDE
plt.figure(figsize=(8,5))
sns.histplot(data["hours-per-week"], bins=30, kde=True)
plt.title("Distribution of Hours per Week")
plt.xlabel("Hours per Week")
plt.ylabel("Frequency")
plt.show()
```

Distribuția este asimetrică la dreapta.

Există un vârf pronunțat în jurul valorii 40, sugerând norma standard de muncă.

Se observă outlieri evidenți (ex. >80 ore/săptămână).

```python
data.describe()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=data["hours-per-week"])
plt.title("Boxplot of Hours per Week")
plt.show()
```

Outlieri clari peste ~60 ore/săptămână

Nu sunt neapărat erori → pot reflecta muncă intensivă sau ocupații specifice

```python
num_cols = ["age", "education-num", "capital-gain"]

for col in num_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

age: distribuție aproape normală, ușor skewed

education-num: discretă, multimodală

capital-gain: extrem de skewed, majoritatea valorilor sunt 0

```python
cat_cols = ["workclass", "education", "sex"]

for col in cat_cols:
    plt.figure(figsize=(10,4))
    data[col].value_counts().plot(kind="bar")
    plt.title(f"Frequency of {col}")
    plt.ylabel("Count")
    plt.show()
```

Private domină categoria workclass

HS-grad, Some-college și Bachelors sunt cele mai frecvente niveluri educaționale

Dezechilibru de gen: mai mulți bărbați decât femei

```python
corr = data.select_dtypes(include=np.number).corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Corelații slabe–moderate:

education-num → pozitiv

age → ușor pozitiv

fnlwgt nu are relevanță explicativă directă

Concluzii prinicipale:

Variabila target hours-per-week este puternic concentrată în jurul valorii de 40, indicând un model dominant de muncă full-time

Distribuția este asimetrică, cu outlieri semnificativi peste 80 ore

Valorile lipsă apar în special în variabile categorice și trebuie tratate înainte de modelare

Variabilele capital-gain și capital-loss sunt extrem de dezechilibrate → transformare necesară

Corelațiile liniare cu target-ul sunt slabe → justifică utilizarea unor modele neliniare
