```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('train_preprocessed.csv')

# Descriptive statistics
print("Descriptive Statistics:")
print(data.describe())

# Missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Histograms for numerical columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols[:3]:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True)
    plt.title(f"Histogram of {col}")
    plt.show()

# Bar plots for categorical columns
categorical_cols = [col for col in data.columns if col not in numerical_cols]
for col in categorical_cols[:3]:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=data[col])
    plt.title(f"Bar Plot of {col}")
    plt.show()

# Outlier detection with boxplots
for col in numerical_cols[:1]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

# Correlation matrix for 'hours-per-week'
if 'hours-per-week' in data.columns:
    correlation_matrix = data.corr()[['hours-per-week']].sort_values(by='hours-per-week', ascending=False)
    plt.figure(figsize=(10, 20))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0)
    plt.title("Correlation of Hours-Per-Week with Other Variables")
    plt.show()
```
