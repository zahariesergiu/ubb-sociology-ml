#### Identify outliers in at least one numerical column using boxplots.

```python
# Importarea librăriilor
import matplotlib.pyplot as plt
import seaborn as sns

# Selectarea coloanei pentru analiză
col = "capital-gain"

# Crearea boxplot-ului
plt.figure(figsize=(8, 5))
sns.boxplot(x=data[col], color="lightblue")
plt.title(f'Boxplot of {col}')
plt.xlabel(col)
plt.show()

# Calcularea outlierilor folosind IQR
Q1 = data[col].quantile(0.25)  # Primul quartil (25%)
Q3 = data[col].quantile(0.75)  # Al treilea quartil (75%)
IQR = Q3 - Q1  # Intervalul interquartilic

# Definirea limitelor pentru outlieri
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificarea valorilor care depășesc limitele
outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)][col]

# Afișarea numărului de outlieri și câteva exemple
print(f"Number of outliers in {col}: {len(outliers)}")
print(outliers.sample(10))  # Afișarea unui eșantion de outlieri
```

Acest cod identifică outlieri (valori extreme) în coloana numerică "capital-gain" folosind boxplot-uri și metoda IQR (Interquartile Range). Se creează mai întâi un boxplot cu Seaborn pentru a vizualiza distribuția datelor. Apoi, se calculează primul cuartil (Q1 - 25%) și al treilea cuartil (Q3 - 75%), iar intervalul intercuartilic (IQR = Q3 - Q1) este determinat. Pe baza acestuia, se stabilesc limitele inferioare și superioare pentru outlieri (lower_bound = Q1 - 1.5 * IQR, upper_bound = Q3 + 1.5 * IQR). Orice valoare din dataset care depășește aceste limite este considerată outlier. În final, codul afișează numărul total de outlieri și un eșantion de 10 astfel de valori. 

Rezultatele boxplot-ului pentru capital-gain arată că această variabilă are un număr semnificativ de outlieri (2712 valori). Majoritatea datelor sunt concentrate aproape de zero, în timp ce valorile extreme sunt dispersate pe intervale mari, ajungând până la 99.999. Acest lucru sugerează că doar un număr mic de persoane au câștiguri de capital semnificative, iar restul nu au deloc sau au valori foarte mici. Outlierii detectați includ valori precum 15.024, 7.298, 3.137, care sunt semnificativ mai mari decât majoritatea punctelor de date. Această distribuție puternic asimetrică indică faptul că majoritatea observațiilor sunt zero, iar câteva valori extreme ar putea influența analiza statistică sau modelele de machine learning. O posibilă abordare ar fi tratarea outlierilor prin scalare robustă.

#### Scalare robustă

```python
# Importarea librăriilor
from sklearn.preprocessing import RobustScaler
import pandas as pd
```

```python
# Aplic scalarea pe coloana capital-gain
scaler = RobustScaler()
data['capital-gain-scaled'] = scaler.fit_transform(data[['capital-gain']])
```

```python
# Verificare
print(data['capital-gain-scaled'].describe())
```

Numărul de observații rămâne același (32.561), ceea ce indică faptul că niciun rând nu a fost eliminat.Mai mult, media și deviația standard sunt aceleași ca în datele originale, ceea ce sugerează că scalarea nu a modificat distribuția datelor, ci doar a ajustat unitatea de măsură.
Minimul și percentila 50% sunt 0, ceea ce confirmă că majoritatea valorilor capital-gain erau zero înainte de scalare.
Maximul rămâne 99.999, ceea ce sugerează că există în continuare valori extreme, dar acum sunt normalizate față de restul datelor.

#### Use a heatmap to visualize correlations among variables and highlight the top three strongest correlations

```python
# Importarea librăriilor
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Selectarea doar a coloanelor numerice
numerical_data = data.select_dtypes(include=["float64", "int64"])

# Calcularea matricei de corelație
correlation_matrix = numerical_data.corr()

# Crearea heatmap-ului
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Correlation Matrix")
plt.show()

# Identificarea celor mai puternice 3 corelații (în valoare absolută, excluzând 1.0)
correlation_pairs = correlation_matrix.unstack().abs().sort_values(ascending=False)
strongest_correlations = correlation_pairs[correlation_pairs < 1.0].drop_duplicates().head(3)

print("Top 3 cele mai puternice corelații:")
print(strongest_correlations)
```

Acest cod analizează corelațiile dintre variabilele numerice din dataset și identifică cele mai puternice 3 corelații (în valoare absolută, excluzând auto-corelațiile de 1.0). Mai întâi, se selectează doar coloanele numerice folosind data.select_dtypes(), apoi se calculează matricea de corelație cu numerical_data.corr(). Se creează un heatmap (sns.heatmap()) pentru a vizualiza relațiile dintre variabile, folosind schema de culori "coolwarm" pentru evidențierea valorilor ridicate și scăzute. Apoi, se extrag perechile de corelații, se iau valorile absolute (abs()), iar acestea sunt sortate descrescător. Se filtrează doar corelațiile mai mici decât 1.0 (correlation_pairs < 1.0) pentru a evita auto-corelațiile și se păstrează doar cele mai puternice 3 corelații distincte (drop_duplicates().head(3)).


Aceast heatmap și lista celor mai puternice corelații arată relațiile dintre variabilele numerice din dataset. Observăm că nu există corelații foarte puternice între variabile, întrucât cele mai mari valori rămân sub 0.15. Cele mai puternice 3 corelații identificate sunt:

education-num și hours-per-week (0.148) – indicând că persoanele cu un nivel mai ridicat de educație tind să lucreze mai multe ore pe săptămână.
education-num și capital-gain (0.123) – sugerând că persoanele cu un nivel educațional mai mare au șanse mai mari de a obține câștiguri de capital.

#### Create a correlation matrix for numerical variables and discuss the strongest and weakest correlations.


```python
# Importarea librăriilor
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Încarc datele
X_train = pd.read_csv("C:/Users/Teodora/Downloads/curs1/ubb-sociology-ml/final_project/X_train.csv")

# Selectez doar coloanele numerice
numerical_X_train = X_train.select_dtypes(include=['number'])

# Calculez  matricea de corelație pentru variabilele numerice
correlation_matrix = numerical_X_train.corr()

# Afișez matricea de corelație sub formă de heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Matricea de corelație pentru variabilele numerice")

# Afișez graficul
plt.show()

# Identific cele mai puternice și cele mai slabe corelații
strongest_correlations = correlation_matrix.unstack().sort_values(ascending=False).drop_duplicates()
weakest_correlations = strongest_correlations[strongest_correlations < 1].tail(5)  # Cele mai slabe corelații
strongest_correlations = strongest_correlations[strongest_correlations < 1].head(5)  # Cele mai puternice corelații

# Afișez rezultatele
# Afișez cele mai puternice corelații
print("Cele mai puternice corelații:")
print(strongest_correlations.head(5))

# Afișez cele mai slabe corelații
print("\nCele mai slabe corelații:")
print(weakest_correlations.tail(5))

```

Acest cod încarcă setul de date X_train.csv, selectează doar variabilele numerice și calculează matricea de corelație dintre acestea. Rezultatul este afișat sub formă de heatmap folosind seaborn.heatmap(), unde corelațiile sunt evidențiate cu o scală de culori pentru o interpretare vizuală mai clară. Apoi, codul identifică cele mai puternice și cele mai slabe corelații dintre variabile, sortându-le și eliminând duplicatele. La final, sunt afișate în consolă primele 5 cele mai puternice corelații și primele 5 cele mai slabe corelații, permițând astfel o analiză rapidă a relațiilor dintre variabilele numerice din dataset.

Matricea de corelație afișează relațiile dintre variabilele numerice din dataset, cu valori cuprinse între -1 și 1. Valorile apropiate de 1 indică o corelație pozitivă puternică, în timp ce valorile apropiate de -1 indică o corelație negativă puternică. În acest caz, nu există corelații puternice între variabile, deoarece toate coeficienții de corelație sunt relativ mici.
