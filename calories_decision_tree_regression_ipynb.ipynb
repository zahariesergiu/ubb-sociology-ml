```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor, plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dataset
weights = np.array([10, 24, 40, 52, 60, 77, 82, 100, 120, 130, 145, 155, 170, 180, 200]).reshape(-1, 1)
calories = np.array([148, 154, 167, 176, 176, 191, 203, 207, 227, 230, 245, 251, 259, 290, 503])
```

```python
# Dataset
weights = np.array([10, 24, 40, 52, 60, 77, 82, 100, 120, 130, 145, 155, 170, 180, 200]).reshape(-1, 1)
calories = np.array([148, 154, 167, 176, 176, 191, 203, 207, 227, 230, 245, 251, 259, 290, 503])

# Train a Decision Tree Regressor
regressor = DecisionTreeRegressor(max_depth=3, random_state=42)
regressor.fit(weights, calories)

# Predicting values (for visualization)
weights_grid = np.arange(min(weights), max(weights), 0.1).reshape(-1, 1)
predicted_calories = regressor.predict(weights_grid)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(weights, calories, color='red', label='Original Data')
plt.plot(weights_grid, predicted_calories, color='blue', label='Decision Tree Regression')
plt.title('Decision Tree Regression')
plt.xlabel('Weights')
plt.ylabel('Calories')
plt.legend()
plt.show()
```

```python
# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(regressor, feature_names=["Weight"], filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree for Regression")
plt.show()
```

## Split for root node

```python
import numpy as np
import pandas as pd

# Dataset
weights = weights.reshape(-1, 1)
# calories = np.array([148, 154, 167, 176, 176, 191, 203, 207, 227, 230, 245, 251, 259, 290, 503])

# Function to compute MSE
def compute_mse(values):
    if len(values) == 0:
        return 0  # Avoid division by zero for empty splits
    mean = np.mean(values)
    mse = np.mean((values - mean) ** 2)
    return mse

# Calculate all possible split points
split_points = (weights[:-1] + weights[1:]) / 2

# Calculate metrics for each split
split_data = []
for split in split_points:
    left_indices = weights.ravel() <= split
    right_indices = weights.ravel() > split

    left_calories = calories[left_indices]
    right_calories = calories[right_indices]

    left_mse = compute_mse(left_calories)
    right_mse = compute_mse(right_calories)

    total_mse = (len(left_calories) * left_mse + len(right_calories) * right_mse) / len(calories)

    split_data.append({
        "Split Point": split[0],
        "Left Count": len(left_calories),
        "Right Count": len(right_calories),
        "Left MSE": left_mse,
        "Right MSE": right_mse,
        "Total MSE": total_mse
    })

# Convert to DataFrame for better visualization
split_df = pd.DataFrame(split_data)
split_df
```
