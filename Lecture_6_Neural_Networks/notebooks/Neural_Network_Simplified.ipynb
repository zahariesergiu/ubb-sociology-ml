### Gradient descent

```python
knob_weight = 0.5
input = 0.5
goal_prediction = 0.8

for iteration in range(20):
	prediction = input * knob_weight
	error = (goal_prediction - prediction) ** 2
	direction_and_amount = (goal_prediction - prediction) * input
	knob_weight = knob_weight + direction_and_amount

	print (f"Error: {error} Prediction: {prediction}")
# The direction of the update is determined by the negative gradient of the cost function with respect to the parameters.
```

### Derivative and play around

```python
knob_weight = 0.5
input = 0.5
goal_prediction = 0.8

for iteration in range(20):
	prediction = input * knob_weight
	error = (goal_prediction - prediction) ** 2
	derivative = input * (goal_prediction - prediction)
	knob_weight = knob_weight + derivative

	print (f"Error: {error} Prediction: {prediction}")
```

### Learning rate

```python
knob_weight = 0.5
input = 2
goal_prediction = 0.8
alpha = 0.1 # make it crazy small or crazy big. What happens? 10 > 0.0001

for iteration in range(20):
	prediction = input * knob_weight
	error = (goal_prediction - prediction) ** 2
	derivative = input * (prediction - goal_prediction)
	knob_weight = knob_weight - (alpha * derivative)

	print (f"Error: {error} Prediction: {prediction}")

 # The learning rate, a hyperparameter, controls the size of the steps taken during each iteration.
```
